### 과목 리스트에서 동일 과목으로 간주되는 수업들을 들은 학생들의 수강 목록을 처리함

- 중복과목 코드들을 하나로 합쳐서 임시코드를 부여

- 학생 수강 목록에서 중복코드에 해당되는 수업에 임시코드를 붙여주기

- 한 학생의 데이터 내에서 동일한 임시코드가 존재한다면 가장 높은 성적을 제외한 다른 중복 수업들을 지우기

```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_excel('master_not_null.xlsx')      #중복코드가 null이 아닌 과목만 가져옴 --> 중복되는 코드들만 묶을것이므로
```

```python
# 중복코드 리스트 구하기
same_code = []

for i in range(len(data)):
    same_code.append(data.loc[i]['중복코드'])
```

```python
# 중복 코드의 중복 제거 -- 위에서는 중복 체크 안하고 모두 리스트에 추가한 것이므로
same_code = list(set(same_code))
```

```python
# 중복 코드명으로 딕셔너리 만들기
code_dic = {}
for elem in same_code:
    code_dic[elem] = []
```

```python
# 만약 중복코드의 값이 위에서 만든 same_code의 key에 있다면 교과목이 해당 코드에 포함되는 것임(동일한 과목이므로)
for i in range(len(data)):
    if data.loc[i]['중복코드'] in list(code_dic.keys()):
        code_dic[data.loc[i]['중복코드']].append(data.loc[i]['교과목코드'])
        # 증복 코드 이름으로 된 리스트를 만든다.
        # 리스트에 교과목 코드를 추가한다.
```

```python
code_dic
```

```python
# 교과목 코드로 해놓았던 key를 임시 코드로 변경
new_keys = []
for i in range(len(code_dic.items())):
    new_keys.append(i)
```

```python
for key,n_key in zip(code_dic.keys(), new_keys):
    code_dic[n_key] = code_dic.pop(key)
```

```python
# 개인별 학생 수강 데이터 불러오기
student_data = pd.read_excel('개인별.xlsx')
```

```python
# 영문학과와 철학과 학생들만 추출
student_data = student_data[(student_data['학과']=='영어영문학과') | (student_data['학과']=='철학과')]
```

```python
# 일단 임시코드 처리 전 모두 -1로 만듬
# 임시코드가 -1이라면 수업 코드가 딱 하나 존재하는 수업들임
student_data['임시코드'] = -1
```

```python
student_data
```

```python
# 위에서 만든 code_dic에 따라 임시코드 부여
for i in range(len(student_data)):
    for k, v in code_dic.items(): #mydict에 아이템을 하나씩 접근해서, key, value를 각각 k, v에 저장
        if student_data.iloc[i]['과목코드'] in v:
            student_data.iloc[i] = student_data.iloc[i].replace(-1,k)
            break
```

```python
# 나중에 엑셀로 export하기 위한 함수
def append_df_to_excel(filename, df, sheet_name='Sheet1', startrow=None,
                       truncate_sheet=False):
    """
    Append a DataFrame [df] to existing Excel file [filename]
    into [sheet_name] Sheet.
    If [filename] doesn't exist, then this function will create it.

    Parameters:
      filename : File path or existing ExcelWriter
                 (Example: '/path/to/file.xlsx')
      df : dataframe to save to workbook
      sheet_name : Name of sheet which will contain DataFrame.
                   (default: 'Sheet1')
      startrow : upper left cell row to dump data frame.
                 Per default (startrow=None) calculate the last row
                 in the existing DF and write to the next row...
      truncate_sheet : truncate (remove and recreate) [sheet_name]
                       before writing DataFrame to Excel file
      to_excel_kwargs : arguments which will be passed to `DataFrame.to_excel()`
                        [can be dictionary]

    Returns: None
    """
    from openpyxl import load_workbook

    import pandas as pd

    writer = pd.ExcelWriter(filename, engine='openpyxl')

    # Python 2.x: define [FileNotFoundError] exception if it doesn't exist 
    try:
        FileNotFoundError
    except NameError:
        FileNotFoundError = IOError


    try:
        # try to open an existing workbook
        writer.book = load_workbook(filename)

        # get the last row in the existing Excel sheet
        # if it was not specified explicitly
        if startrow is None and sheet_name in writer.book.sheetnames:
            startrow = writer.book[sheet_name].max_row

        # truncate sheet
        if truncate_sheet and sheet_name in writer.book.sheetnames:
            # index of [sheet_name] sheet
            idx = writer.book.sheetnames.index(sheet_name)
            # remove [sheet_name]
            writer.book.remove(writer.book.worksheets[idx])
            # create an empty sheet [sheet_name] using old index
            writer.book.create_sheet(sheet_name, idx)

        # copy existing sheets
        writer.sheets = {ws.title:ws for ws in writer.book.worksheets}
    except FileNotFoundError:
        # file does not exist yet, we will create it
        pass

    if startrow is None:
        startrow = 0

    # write out the new sheet
    df.to_excel(writer, sheet_name, startrow=startrow)

    # save the workbook
    writer.save()
```

```python
# 학생별로 데이터 나누기(기준: 학번)
gb = student_data.groupby('학번')    
personal_data = [gb.get_group(x) for x in gb.groups]


# 차례로 중복 제거
def remove_dup(df):
    df_1 = df[df['임시코드'] != -1]
    df_2 = df[df['임시코드'] == -1]  # with minus

    df_1 = df_1.drop_duplicates(['임시코드'], keep='first') #중복이 발생한다면 학점이 높은 수업 하나만 남기고 제거

    final_student_data = pd.concat([df_1, df_2])
    return final_student_data

 
# 학생별로 데이터 처리하고 나중에 합쳐서 excel파일로 export하기
for i in range(len(personal_data)):
    df = personal_data[i]
    df = df.sort_values(by=['등급'], axis=0)
    df = remove_dup(df)
    
    append_df_to_excel('final2.xlsx', df, sheet_name='Sheet1', startrow=None,
                       truncate_sheet=False)
```
